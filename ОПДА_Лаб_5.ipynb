```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.read_csv('Independence.csv')
df
```

# 1.1. Box plots.



```python
sns.boxenplot(df['Sales'])
```

```python
sns.boxplot(x='Average Check',data=df)
```

```python
sns.kdeplot(df['Sales'])
```

# 1.2. Violin plot.

```python
sns.violinplot(x='City', y='Meals Served', data=df[:10])
```

# 1.3. Count plots.

```python
sns.countplot(df['State'])
```

# 1.4. Heatmap.

```python
sns.heatmap(df.corr(),annot=True,cmap='YlGnBu')
```

# 1.5. Facet grid.

```python
g = sns.FacetGrid(df, col="State")
g.map(sns.histplot, "Sales")
```

# 1.6. Regplot.

```python
sns.regplot(x='Sales', y='Average Check', data=df)
```

# 1.7. lmplot.

```python
sns.lmplot(x='Sales',y='Average Check', hue='State', data=df)
```

# 1.8. Swarm plot

```python
sns.swarmplot(x='State', y='Meals Served', data=df[:50])
```

# 1.9. Pair plot

```python
sns.pairplot(data=df)
```

# 1.10. Joint plot

```python
sns.jointplot(x='Sales', y='State', data=df)
```
