```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv('Top250.csv')
df.head()
```

#Обработка данных

```python
df.info()
```

```python
df.drop(['Rank', 'Headquarters', 'Content'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['YOY_Sales']=df['YOY_Sales'].str.replace(r"%", "")
df['YOY_Units']=df['YOY_Units'].str.replace(r"%", "")
```

```python
df['YOY_Sales']= df['YOY_Sales'].astype('float64')
df['YOY_Units']= df['YOY_Units'].astype('float64')
```

```python
df.head()
```

```python
df.info()
```
